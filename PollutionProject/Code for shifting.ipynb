```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'F:\project dataset\SeparateWithMonth.csv')
```

```python
df['PM25_1'] = df.PM25.shift(periods=1)
df['PM10_1'] = df.PM10.shift(periods=1)
df['Temp_1'] = df.Temp.shift(periods=1)
df['RH_1'] = df.RH.shift(periods=1)

df['PM25_2'] = df.PM25.shift(periods=2)
df['PM10_2'] = df.PM10.shift(periods=2)
df['Temp_2'] = df.Temp.shift(periods=2)
df['RH_2'] = df.RH.shift(periods=2)

df['PM25_3'] = df.PM25.shift(periods=3)
df['PM10_3'] = df.PM10.shift(periods=3)
df['Temp_3'] = df.Temp.shift(periods=3)
df['RH_3'] = df.RH.shift(periods=3)

df['PM25_4'] = df.PM25.shift(periods=4)
df['PM10_4'] = df.PM10.shift(periods=4)
df['Temp_4'] = df.Temp.shift(periods=4)
df['RH_4'] = df.RH.shift(periods=4)
```

```python
print(df)
```

```python
df = df.dropna(axis=0,subset=['PM25_1','PM25_2','PM25_3','PM25_4','PM10_1','PM10_2','PM10_3','PM10_4','Temp_1','Temp_2','Temp_3','Temp_4','RH_1','RH_2','RH_3','RH_4' ])
df.shape
```
